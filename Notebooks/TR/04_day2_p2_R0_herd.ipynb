### Vaka Ã§alÄ±ÅŸmasÄ±

4)	Modelinizi Ã§alÄ±ÅŸtÄ±rdÄ±ÄŸÄ±nÄ±zda, aÅŸaÄŸÄ±daki bilgileri verebilir misiniz:   

  a)	Bir yÄ±l sonra salgÄ±nÄ±n beklenen bÃ¼yÃ¼klÃ¼ÄŸÃ¼ nedir?
  b)	Mevcut koÅŸullar altÄ±nda kaÃ§ Ã¶lÃ¼mÃ¼n meydana geleceÄŸi Ã¶ngÃ¶rÃ¼lmektedir?
  c)	SalgÄ±n ne zaman zirve yapacak?
  d)	Bu salgÄ±n hakkÄ±ndaki mevcut bilgi gÃ¶z Ã¶nÃ¼ne alÄ±ndÄ±ÄŸÄ±nda, sÃ¼rÃ¼ baÄŸÄ±ÅŸÄ±klÄ±ÄŸÄ± eÅŸiÄŸi iÃ§in deÄŸer nedir?
  
Not: Soru iÅŸaretlerini (?) arayÄ±n ve gerektiÄŸinde deÄŸiÅŸtirin.


```python
# PAKETLERÄ° YÃœKLE:
library(deSolve)
library(reshape2)
library(ggplot2)
library(here)

setwd(here())
# PopÃ¼lasyon bÃ¼yÃ¼klÃ¼ÄŸÃ¼:
N <- 100000

# MODEL GÄ°RÄ°ÅLERÄ°:

# BaÅŸlangÄ±Ã§ koÅŸullarÄ±
initial_state_values <- c(
  S = ??    # <------------S iÃ§in baÅŸlangÄ±Ã§ koÅŸullarÄ±nÄ± yaz
  E = 0,
  I = ??    # <----------- I iÃ§in baÅŸlangÄ±Ã§ koÅŸullarÄ±nÄ± yaz
  R = 0,
  M = 0)           

# GÃ¼nlÃ¼k parametre deÄŸerleri  <--------------AÅŸaÄŸÄ±daki model parametrelerini tamamla

epsilon <- ??       # semptom baÅŸlangÄ±cÄ± oranÄ±  
gamma   <- ??       # Ä°yileÅŸme oranÄ± 
R0      <- ??       # R0
beta    <- ??       # enfeksiyon oranÄ±
cfr     <- ??       # Vaka Ã¶lÃ¼m oranÄ±
mu      <- ??       # Ã–lÃ¼m oranÄ± [rekabetÃ§i tehlikeleri hatÄ±rlayÄ±n! gamma*CFR/(1-CFR) ]


parameters <- c(beta,epsilon, gamma, mu)


# MODEL FONKSÄ°YONU: 
times <- seq(from = 0, to = 365, by = 1)   # 0'dan 365 gÃ¼ne kadar gÃ¼nlÃ¼k aralÄ±klarla

seir_model <- function(time, state, parameters) {  
  
  with(as.list(c(state, parameters)), {     
    
    # Toplam popÃ¼lasyon bÃ¼yÃ¼klÃ¼ÄŸÃ¼ N'yi hesaplama (her bÃ¶lmedeki kiÅŸi sayÄ±sÄ±nÄ±n toplamÄ±)
    
    N <- ??  # <----------------- toplam popÃ¼lasyonu tanÄ±mla
    
    # FOI'yi hesaplama
    
    lambda <- beta*I/N  #  Enfeksiyon kuvvetini nasÄ±l tanÄ±mlayabiliriz? 
    
    # Diferansiyel denklemler
    dS <-  -lambda*S 
    dE <-   ??                  # E iÃ§in diferansiyel denklemi yazabilir misin? 
    dI <-   E*epsilon - I*(gamma+mu)     
    dR <-   gamma * I 
    dM <-   I*mu
    
    return(list(c(dS, dE, dI, dR, dM))) 
  })
  
}


# MODEL Ã‡IKTISI (diferansiyel denklemleri Ã§Ã¶zme):

# ode entegrasyon algoritmasÄ±nÄ± kullanarak diferansiyel denklemleri Ã§Ã¶zme
output <- as.data.frame(ode(y = initial_state_values, 
                            times = times, 
                            func = seir_model,
                            parms = parameters))



# Ã§Ä±ktÄ± veri setini uzun formata Ã§evir
output_long <- melt(as.data.frame(output), id = "time")                 


# S, I ve R bÃ¶lmelerindeki kiÅŸi sayÄ±sÄ±nÄ± zaman iÃ§inde Ã§izdir
ggplot(data = output_long,                                               # Ã§izilecek veriyi iÃ§eren nesneyi belirt
       aes(x = time, y = value, colour = variable, group = variable)) +  # sÃ¼tunlarÄ± eksenlere ve gruplara ata
  geom_line(size=2) +                                                          # veriyi Ã§izgiler olarak gÃ¶ster
  xlab("Zaman (gÃ¼n)")+                                                   # x ekseni iÃ§in etiket ekle
  ylab("SayÄ±") +                                                       # y ekseni iÃ§in etiket ekle
  labs(colour = "BÃ¶lme",                                           # lejand baÅŸlÄ±ÄŸÄ± ekle  
       title = "COVID-19 (SEIR) modeli")    
```
